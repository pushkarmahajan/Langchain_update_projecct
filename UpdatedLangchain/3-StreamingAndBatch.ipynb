### Streaming and Batch

### Streaming

Most models can stream their output content while it is being generated. By displaying output progressively, streaming significantly improves user experience, particularly for longer response. Calling stream() returns an iterator that yields output chunks as they are produced. You can use a loop to process each chunk in real-time.

```python
import os
from dotenv import load_dotenv
from langchain.chat_models import  init_chat_model
load_dotenv()

os.environ["GoogleAPIKey"] = os.getenv("googleAIStudioKey")
os.environ["GROQ_API_KEY"] = os.getenv("groqAPIKey")

model = init_chat_model("groq:llama-3.1-8b-instant")

# how to invoke a model
response = model.invoke("Hello how are you?")
response.content
```

```python
for chunk in model.stream("Write me a 200 words paragraph on Artificial Intelligencce"):
    print(chunk.text, end="||", flush=True)
```

### Batch

Batching a collection of independent requests to a model can significantly improve performance and reduce costs, as the processing can be done in parallel:

```python
responses = model.batch([
"How to loose fat in 3 days",
"Why do parrots have colourful beaks",
"How to make buttermilk",
],
config={ "max_concurrency" : 2,}
)
for response in responses:
    print(response)
```
